```python
def find_groups_with_types(df, required_types):
    required_types = set(required_types)

    result = (
        df.groupby(["field_id", "crop_name"])["historical_values.product_type"]
        .agg(lambda x: required_types.issubset(set(x.dropna())))
        .reset_index(name="has_required_types")
    )

    return result.loc[
        result["has_required_types"],
        ["field_id", "crop_name"],
    ].reset_index(drop=True)
```

```python
required = [
    "ndvi",
    "ndvi_ps8",
    "ndvi_s2a",
    "ndvi_s2b",
    "soil_moisture",
    "temperature",
    "ndvi_s2c",
    "ndvi_l8",
]

groups = find_groups_with_types(df, required)

groups
```

```python
def leave_one_out_group_counts(df, required_types):
    result = []

    for removed_type in required_types:
        current_required = [t for t in required_types if t != removed_type]

        groups = find_groups_with_types(df, current_required)

        result.append(
            {
                "removed_type": removed_type,
                "remaining_required": current_required,
                "n_groups": len(groups),
            }
        )

    return pd.DataFrame(result).sort_values("n_groups", ascending=False)


counts = leave_one_out_group_counts(df, required)

display(counts)
```

```python
counts_by_type = (
    df.groupby("historical_values.product_type")[["field_id", "crop_name"]]
    .apply(lambda x: x.drop_duplicates().shape[0])
    .reset_index(name="n_groups")
    .sort_values("n_groups", ascending=False)
)

display(counts_by_type)
```

```python
tmp = df.copy()

tmp["value_len"] = tmp["historical_values.value"].apply(len)

stats_by_type = (
    tmp.groupby("historical_values.product_type")
    .agg(
        mean_len=("value_len", "mean"),
        min_len=("value_len", "min"),
        max_len=("value_len", "max"),
    )
    .reset_index()
)

group_counts = (
    tmp[["field_id", "crop_name", "historical_values.product_type"]]
    .drop_duplicates()
    .groupby("historical_values.product_type")
    .size()
    .reset_index(name="n_groups")
)

stats_by_type = stats_by_type.merge(
    group_counts, on="historical_values.product_type"
).sort_values("n_groups", ascending=False)

display(stats_by_type)
```

```python
import json
import numpy as np
import pandas as pd
import torch

from torch.utils.data import Dataset


class YieldDataset(Dataset):
    def __init__(
        self,
        df,
        features,
        seq_len=20,
        step_days=3,
    ):
        self.features = features
        self.seq_len = seq_len
        self.step_days = step_days

        df = df.copy()

        df["sowing_date"] = pd.to_datetime(df["sowing_date"])
        df["harvesting_date"] = pd.to_datetime(df["harvesting_date"])

        df["historical_values.value"] = df["historical_values.value"].apply(json.loads)

        self.samples = []

        for _, group in df.groupby(["field_id", "crop_name"]):
            first = group.iloc[0]

            start = first["sowing_date"]
            end = first["harvesting_date"]

            feature_data = {}

            for _, row in group.iterrows():
                feature = row["historical_values.product_type"]

                if feature not in features:
                    continue

                values = pd.DataFrame(
                    row["historical_values.value"],
                    columns=["date", "value"],
                )

                values["date"] = pd.to_datetime(values["date"])

                feature_data[feature] = values

            current_date = start + pd.Timedelta(days=seq_len * step_days)

            while current_date <= end:
                x = []

                for i in range(seq_len):
                    t = current_date - pd.Timedelta(days=(seq_len - i) * step_days)

                    row_values = []

                    for feature in features:
                        value = 0.0
                        valid = 0.0

                        if feature in feature_data:
                            values = feature_data[feature]

                            idx = (values["date"] - t).abs().idxmin()

                            nearest = values.loc[idx]

                            delta_days = abs((nearest["date"] - t).days)

                            if delta_days <= step_days:
                                v = nearest["value"]
                                if isinstance(v, list):
                                    value = float(np.mean(v))
                                else:
                                    value = float(v)
                                valid = 1.0

                        row_values.extend([value, valid])

                    x.append(row_values)

                self.samples.append(
                    {
                        "x": np.array(x, dtype=np.float32),
                        "y": np.float32(first["productivity"]),
                        "field_id": first["field_id"],
                        "crop_name": first["crop_name"],
                        "prediction_date": current_date,
                    }
                )

                current_date += pd.Timedelta(days=step_days)

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]

        return {
            "x": torch.tensor(sample["x"]),
            "y": torch.tensor(sample["y"]),
            "field_id": sample["field_id"],
            "crop_name": sample["crop_name"],
            "prediction_date": sample["prediction_date"],
        }
```

```python
features = [
    "ndvi",
    "ndvi_ps8",
    "ndvi_s2a",
    "ndvi_s2b",
    "soil_moisture",
    "temperature",
]

dataset = YieldDataset(
    df=df,
    features=features,
    seq_len=20,
    step_days=3,
)

print(len(dataset))
```
